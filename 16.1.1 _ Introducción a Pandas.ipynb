# Añadiendo una columna nueva

```python
# Se hará el mismo tratamiento que se hizo en lecciones anteriores, para
# generar la duración de cada registro
# Como primer paso, se añadirá un nuevo campo de duración, para no modificar la estructura original

df["duration_num"] = df["duration"]

# Además se generará un campo adicional de unidades, para guardar la info., que podrá ser "min" o "season".
df["duration_unit"] = ""

# Se verifica el nuevo campo
df.head(3)
```

```python
# Como primer paso, se cambia el tipo de la columna de las columnas de duration
df["duration_unit"] = df["duration_unit"].astype(pd.StringDtype())

# Si lo quisiera hacer para más columnas a la vez (todas las de tipo object)
# Podría primero buscar todas las columnas de ese tipo
df.select_dtypes(include="object").columns

# Y luego aplicarlo
```

```python
df.dtypes
```

```python
# Cambia el tipo de variables de todas las columnas tipo object
stringcols = df.select_dtypes(include="object").columns
df[stringcols] = df[stringcols].fillna("").astype(pd.StringDtype())

df.dtypes
```
