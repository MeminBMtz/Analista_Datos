```python
arreglo = np.array([["Bella", "Labrador", "Café", 56, 24, "2013-07-01"],
                   ["Charlie", "Poodle", "Negro", 43, 24, "2016-09-16"],
                   ["Lucy", "Chow Chow", "Café", 46, 24, "2014-08-25"],
                   ["Cooper", "Schnauzer", "Gris", 49, 17, "2011-12-11"],
                   ["Bernie", "San Bernardo", "Blanco", 77, 74, "2018-02-27"]])
arreglo
```

```python
df = pd.DataFrame(data = arreglo, columns = ["Nombre", "Raza", "Color", "Altura_cm", "Peso_kg", "Fecha_Nac"])
df
```

```python
# Inciso 1
df2 = df[df["Fecha_Nac"] < "2015-01-01"]
df2
```

```python
# Inciso 2
df["Altura_cm"] = df["Altura_cm"].astype(int) # Convierte los valores de una columna a enteros
```

```python
df3 = df[df["Altura_cm"] > 50]
df3
```

```python
# Inciso 3
df["Peso_kg"] = df["Peso_kg"].astype(int) # Convierte todos los valores de peso a enteros

df["IMM"] = df["Peso_kg"] / ((df["Altura_cm"] / 100) ** 2)
df
```

```python
# Otras manipulaciones...
# Extracción de las columnas 'Nombre', 'Raza' e 'IMM'
df4 = df[["Nombre", "Raza", "IMM"]]
df4
```

```python
df4 = df4.sort_values("IMM", ascending = True)
df4
```
