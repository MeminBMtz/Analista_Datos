```python
import numpy as np
import pandas as pd

# Base de datos de iris
iris = pd.read_csv("Iris.csv")

# Escogemos solo valores cuyo ancho del sépalo sea >= 3.5 y largo del sépalo > 5
condicion1 = iris["Sepal_width"] >= 3.5
condicion2 = iris["Sepal_length"] > 5
df = iris[condicion2 & condicion1]
df
```

### Inciso 1: tabla de medidas estadísticas

```python
inciso1 = df.groupby("Species_name")["Petal_width"].describe()
inciso1[["mean", "std", "50%"]]
```

### Inciso 2: registros del df anterior

```python
# Inciso 2
print(f"Registros para obtener resultados: {len(df)}.")
```

### Inciso 3: estandarización del ancho de cada registro del DF 1

```python
df["petal.width.st"] = (df["Petal_width"] - df["Petal_width"].mean()) / df["Petal_width"].std()
```

```python
df
```

### Inciso 4: Cambio de nombre de columnas "variety" ("Species_name") y "petal.width.st"

```python
df.rename(columns = {"Species_name": "Variedad", "petal.width.st": "Ancho Est. Pétalo"}, inplace = True)
df
```

### Inciso 5: Exportación del DF como archivo .csv

```python
df.to_csv("Iris2.csv")
```
