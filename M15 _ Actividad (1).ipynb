# Uso de BeautifulSoup y requests para parsear info.

```python
from bs4 import BeautifulSoup
from pprint import pprint

# Url es el sitio web que se va a scrapear
url = "https://listado.mercadolibre.com.mx/books#D[A:books]"

headers = {
    "Access-Control-Allow-Origin": "*",
    "Access-Control-Allow-Methods": "GET",
    "Access-Control-Allow-Headers": "Content-Type",
    "Access-Control-Max-Age": "3600",
    "User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0"
}

data = requests.get(url, headers=headers)

html = BeautifulSoup(data.text, 'html.parser')

# Impresión del contenido con BS y requests
for item in html.select('li.ui-search-layout__item'):
    try:
        print('---------------------------')
        title = item.select('h2')[0].get_text()
        
        print(title)
    except Exception as e:
        # raise e
        print('')
```

# Listado simple con info. de los ítems (libros) del sitio web

```python
# Vamos a juntar título, precio y rating
import numpy as np
arr_titles = []
arr_prices = []
arr_rating = []

for item in html.select('li.ui-search-layout__item'):
    try:
        title = item.select("h2.ui-search-item__title")[0].get_text()
        price = item.select("span.andes-money-amount__fraction")[0].get_text()
        
        # Hay ítems sin rating, vamos a manejar esto. Si tiene rating, lo guardamos
        if len(item.select("span.ui-search-reviews__rating-number")) != 0:
            rating = item.select("span.ui-search-reviews__rating-number")[0].get_text()
        else:
            # Si no tiene rating, item.select(...) será una lista vacía, lo que provocará un error
            # por eso le asignamos rating = nan (no tiene)
            rating = np.NaN
        
        arr_titles += [title]
        arr_prices += [price]
        arr_rating += [rating]
    except Exception as e:
        raise e

# Imprimimos los 10 primeros libros, quitamos las comas
for i in range(10):
    if "," in arr_titles[i]:
        new = arr_titles[i].replace(",", "-"); arr_titles[i] = new
    print("-----------------------------------")
    print(f"TÍTULO: {arr_titles[i]}\nPRECIO: ${arr_prices[i]}\nRATING: {arr_rating[i]}")
```

# Uso de requests con respuestas de API

```python
# Para obtener frases aleatorias de Breaking Bad
url = "https://api.breakingbadquotes.xyz/v1/quotes/5"

# Se genera una función simple llamada "jprint"
import json
def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4); print(text)

quotes = requests.get(url)

if quotes.status_code == 200: jprint(quotes.json())
```
