# Lectura y muestra (20 primeras filas) del archivo

```python
import pandas as pd

# Lectura del archivo
df = pd.read_csv('fifa_eda.csv')

# Primeras 20 filas
df.head(20)
```

# Programa anterior

```python
# Código anterior
import numpy as np

# Transformación de la columna de contrato
years = []
for year in df["Contract Valid Until"]:
    if type(year) == str: years.append(int(year[:4])) # Es un valor entero
    else: years.append('nan') # Es float (nan)

from statistics import mode
most_common_year = mode(years) # Año más frecuente

for i in range(len(years)):
    if years[i] == 'nan': # Iteramos la lista por índice y si el valor es nan se asigna el año más frecuente
        years[i] = most_common_year

# Cambiamos los valores del dataset
df['Contract Valid Until'] = np.array(years)
df['Contract Valid Until'].astype('int64') # Los volvemos numéricos de tipo int64

# Nueva columna
current_year = 2024
df["Years Playing"] = current_year - df["Joined"]

# Jugadores de México
mexico_players = df.loc[df['Nationality'] == 'Mexico']

# Ordenados por Release Clause
sorted_release_clause = df.sort_values(by='Release Clause')

# Jugadores por año en que se unieron
new_dataset = df.groupby("Joined").size().rename('Num. Jugadores').reset_index()
```

```python
# Debo cambiar el nombre de algunos países
df.replace('Bosnia Herzegovina', 'Bosnia and Herzegovina', inplace=True)
df.replace('Korea Republic', 'South Korea', inplace=True)
df.replace('Central African Rep.', 'Central African Republic', inplace=True)
df.replace('FYR Macedonia', 'Republic of Macedonia', inplace=True)
df.replace('China PR', 'China', inplace=True)
df.replace('Guinea Bissau', 'Guinea-Bissau', inplace=True)
df.replace('Congo', 'Republic of the Congo', inplace=True)
df.replace('Trinidad & Tobago', 'Trinidad and Tobago', inplace=True)
df.replace('São Tomé & Príncipe', 'São Tomé and Príncipe', inplace=True)
df.replace('Korea DPR', 'North Korea', inplace=True)
df.replace('St Kitts Nevis', 'Saint Kitts and Nevis', inplace=True)
df.replace('Antigua & Barbuda', 'Antigua and Barbuda', inplace=True)
df.replace('St Lucia', 'Saint Lucia', inplace=True)
```

```python
# Primero necesito el código del país
# Para países que no estén en CountryInfo investigué los datos adicionales (capital y zona horaria).
def CountryData(country='England', value='iso'):
    countries = {
        'Montenegro': {'iso': 'ME', 'capital': 'Podgorica', 'timezone': pytz.country_timezones['ME']},
        'China': {'iso': 'CN', 'capital': 'Beijing', 'timezone': pytz.country_timezones['CN']},
        'Kosovo': {'iso': 'XK', 'capital': 'Pristina', 'timezone': pytz.country_timezones['BA']},
        'Gambia': {'iso': 'GM', 'capital': 'Banjul', 'timezone': pytz.country_timezones['GM']},
        'Curacao': {'iso': 'CW', 'capital': 'Willemstad', 'timezone': pytz.country_timezones['CW']},
        'Palestine': {'iso': 'PS', 'capital': 'Ramallah', 'timezone': pytz.country_timezones['PS']},
        'Andorra': {'iso': 'AD', 'capital': 'Andorra la Vella', 'timezone': pytz.country_timezones['AD']},
        'England': {'iso': 'GB-ENG', 'capital': 'London', 'timezone': pytz.country_timezones['GB']},
        'Scotland': {'iso': 'GB-SCT', 'capital': 'Edinburgh', 'timezone': pytz.country_timezones['GB']},
        'Wales': {'iso': 'GB-WLS', 'capital': 'Cardiff', 'timezone': pytz.country_timezones['GB']},
        'Northern Ireland': {'iso': 'GB-NIR', 'capital': 'Belfast', 'timezone': pytz.country_timezones['GB']}
    }
    return countries[country][value]
```

```python
def GetCountry(code='GB-ENG'):
    codes = {
        'ME': 'Montenegro',
        'CN': 'China',
        'XK': 'Kosovo',
        'GM': 'Gambia',
        'CW': 'Curacao',
        'PS': 'Palestine',
        'AD': 'Andorra',
        'GB-ENG': 'England',
        'GB-SCT': 'Scotland',
        'GB-WLS': 'Wales',
        'GB-NIR': 'Northern Ireland'
    }; return codes[code]
```

```python
# Países que no están en CountryInfo
other_countries = ['England', 'Scotland', 'Wales', 'Northern Ireland', # de UK
                   'Montenegro', 'China', 'Kosovo', 'Gambia', 'Curacao', 'Palestine', 'Andorra'] # no de UK
other_countries_codes = ['GB-ENG', 'GB-SCT', 'GB-WLS', 'GB-NIR', # de UK
                        'ME', 'CN', 'XK', 'GM', 'CW', 'PS', 'AD'] # no de UK
# Código ISO
df['Country Code'] = df['Nationality'].apply(lambda x:
                                             CountryInfo(x).info()['ISO']['alpha2'] if x not in other_countries else CountryData(country=x, value='iso'))
# Zona horaria
df['Time Zone'] = df['Country Code'].apply(lambda x:
                                           pytz.country_timezones[x] if x not in other_countries_codes else CountryData(country=GetCountry(code=x), value='timezone'))
# Capital
df['Capital'] = df['Nationality'].apply(lambda x:
                                        CountryInfo(x).capital() if x not in other_countries else CountryData(country=x, value='capital'))
df
```
