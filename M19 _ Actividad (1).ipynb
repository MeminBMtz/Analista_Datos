# Totales

```python
# Total de videojuegos analizados
print('Total de videojuegos analizados:', df['Name'].nunique())
# Total de géneros analizados
print('Total de géneros analizados:', df['Genre'].nunique())
# Total de plataformas analizadas
print('Total de plataformas analizadas:', df['Platform'].nunique())
```

# Ventas

```python
ventas_por_año = df.groupby('Year')['Global_Sales'].sum()
print('Total de ventas por año:')
ventas_por_año
```

```python
# Juegos más vendidos por año
best_sellers_year = df.loc[df.groupby('Year')['Global_Sales'].idxmax(), ['Year', 'Name', 'Global_Sales']]
best_sellers_year
```

```python
# Máximos por plataforma por año
platforms_gross_year = df.groupby(['Year', 'Platform', 'NA_Sales', 'EU_Sales', 'JP_Sales',
                                   'Other_Sales'])['Global_Sales'].sum().reset_index()
# Índices de máximos por año
maximums = platforms_gross_year.groupby('Year')['Global_Sales'].idxmax()
# Plataformas con más ingresos por año
platforms_gross_year.loc[maximums]
```

```python
# Ventas por género y año
sales_genre_year = df.groupby(['Year', 'Genre'])['Global_Sales'].sum().unstack()
sales_genre_year
```

```python
# Gráfico de comparación de ventas entre géneros por año
sales_genre_year.plot(kind='bar', stacked=True, figsize=(10, 5))
plt.title('Comparación de ventas entre géneros por año')
plt.xlabel('Año')
plt.ylabel('Ventas Globales (millones)')
plt.show()
```

# Insights

```python
# Gráfico adicional 1: Comparación de ventas por región y año
ventas_por_region_año = df.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
ventas_por_region_año.plot(kind='line', figsize=(10, 5))
plt.title('Comparación de ventas por región y año')
plt.xlabel('Año')
plt.ylabel('Ventas (millones)')
plt.show()
```

```python
# Gráfico adicional 2: Distribución de ventas globales por género
sns.boxplot(x='Genre', y='Global_Sales', data=df)
plt.title('Distribución de ventas globales por género')
plt.xlabel('Género')
plt.ylabel('Ventas Globales (millones)')
plt.xticks(rotation=90)
plt.show()
```
